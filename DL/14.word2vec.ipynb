```python
import gensim as gen
import pandas as pd
```

```python
df=pd.read_json("/content/Cell_Phones_and_Accessories_5.json",lines=True)
df.head(5)
```

```python
df.shape
```

```python
df.reviewText[0]
```

```python
review_text=df.reviewText.apply(gen.utils.simple_preprocess)
review_text
```

```python
model=gen.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(review_text,progress_per=1000)
```

```python
model.epochs
```

```python
model.corpus_count
```

```python
model.train(review_text,total_examples=model.corpus_count,epochs=5)
```

```python
model.save("amazon.model")
```

```python
model.wv.most_similar("bad")
```

```python
model.wv.similarity(w1="cheap",w2="inexpensive")
```
