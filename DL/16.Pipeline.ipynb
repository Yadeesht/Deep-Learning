```python
def get_label(path):
  return tf.strings.split(path,os.path.sep)[-2]
```

```python
def get(path):
  label=get_label(path)
  img = tf.io.read_file(path)
  img = tf.image.decode_jpeg(img)
  img = tf.image.resize(img,[128,128])
  return img,label
```

```python
train=train.map(get)
for img,label in train.take(3):
  print(img)
  print(label)
```

```python
def scale(image,label):
  return image/255,label
```

```python
train=train.map(scale)
for img,label in train.take(3):
  print(img)
  print(label)
```

```python
reviews_ds=tf.data.Dataset.list_files("/content/comments/*/*",shuffle=False)
```

```python
for i in reviews_ds:
  print(i.numpy())
```

```python
def extract_review(path):
  return tf.io.read_file(path),tf.strings.split(path,os.path.sep)[-2]
```

```python
reviews_ds_1 = reviews_ds.map(extract_review)
for review, label in reviews_ds_1:
    print("Review: ",review.numpy()[:])
    print("Label: ",label.numpy())
```

```python
reviews_ds_2 = reviews_ds_1.filter(lambda review, label: review!="")
for review, label in reviews_ds_2.as_numpy_iterator():
    print("Review: ",review[:])
    print("Label: ",label)
```

```python
final_ds = reviews_ds.map(extract_review).filter(lambda review, label: review!="").shuffle(3)
for review, label in final_ds.as_numpy_iterator():
    print("Review:",review[:])
    print("Label:",label)
```
